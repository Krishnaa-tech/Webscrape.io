```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import pickle
new_df = pd.read_csv('test_classified_articles_data.csv')
new_df
```

```python
with open('trained_model.pkl', 'rb') as model_file:
    trained_model = pickle.load(model_file)

def apply_model_to_new_data(trained_model, new_data):
    # Load the pre-trained vectorizer and classifier
    vectorizer = trained_model['vectorizer']
    classifier = trained_model['classifier']

    # Transform the new text data using the pre-trained vectorizer
    X_new = vectorizer.transform(new_data['Article_Text'])

    # Predict the categories for the new dataset
    y_pred_new = classifier.predict(X_new)

    # Add the predicted categories to the new dataset
    new_data['Predicted_Category'] = y_pred_new

    return new_data


# Apply the pre-trained model to the new dataset
result_df = apply_model_to_new_data(trained_model, new_df)

# Display the result
print(result_df)
```
